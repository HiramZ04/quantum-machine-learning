#Cirq Practices#


```python
import numpy as np
import cirq

q0, q1, q2, q3 = cirq.LineQubit.range(4)

ops = [
    cirq.ry(np.pi/4).on(q0),
    cirq.CNOT(q0, q1),
    cirq.ry(np.pi/4).on(q1),
    cirq.CNOT(q1, q2),
    cirq.ry(np.pi/4).on(q2),
    cirq.CNOT(q2, q3),
    cirq.ry(np.pi/4).on(q3),
]

circuit = cirq.Circuit(ops)
print(circuit)
```
